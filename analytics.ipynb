### Анализ данных, полученных в ходе опроса пользователей интернета, для выявления их отношения к онлайн-играм

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('NewForm.csv')
display(data.head(10))
print('Размерность:', data.shape)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
count = data['Ваш пол'].value_counts()
pie = ax.pie(
    count,
    labels=['Женщины', 'Мужчины'],
    autopct='%1.1f%%',
    startangle=90
)
title = ax.set_title('Соотношение респондентов по гендерному признаку', fontsize=16)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
count = data['Сколько вам лет?'].value_counts() #.sort_values(ascending=False)
pie = ax.pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90
)
title = ax.set_title('Соотношение респондентов по возрасту', fontsize=16)
plt.show(pie)
```

```python
mask = data['Ваш пол'] == 'Женский'
mask2 = data['Ваш пол'] == 'Мужской'
data_tmp = data[mask]
data_tmp2 = data[mask2]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))

count = data_tmp['Сколько вам лет?'].value_counts() 
count2 = data_tmp2['Сколько вам лет?'].value_counts()

axes[0].pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90
)
axes[0].set_title('Соотношение респондентов по возрасту\n среди женщин', fontsize=12);
axes[1].pie(
    count2,
    labels=count2.index,
    autopct='%1.1f%%',
    startangle=90
)

axes[1].set_title('Соотношение респондентов по возрасту\n среди мужчин', fontsize=12);
plt.show(pie)
```

### Вывод.

Видим, что как среди женщин, так и среди мужчин преобладают люди возрастов 25-45 лет (примерно 63% опрошенных), и 18-25 лет (примерно 28% опрошенных).

```python
data['Играете ли вы в онлайн-игры?'].value_counts()
```

Ответы всего двух респондентов на вопрос: 'Играете ли вы в онлайн-игры?' можно считать неоднозначными: 'Да, на компьютере;Нет, не играю' и 'Да, на телефоне;Нет, не играю'. Будем считать их статистически незначимыми.

```python
data_tmp = []
for k,i in enumerate(data['Играете ли вы в онлайн-игры?']):      
    if ';' in i:        
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp.append(j)                 
    else:
        data_tmp.append(i)   
        
data_tmp = pd.Series(data_tmp)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
count = data_tmp.value_counts().sort_values() 
display(count)
pie = ax.pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90, 
    explode = [0, 0, 0.01, 0]
)
title = ax.set_title('Играете ли вы в онлайн-игры?', fontsize=16)
plt.show(pie)
```

```python
mask = data['Ваш пол'] == 'Женский'
mask2 = data['Ваш пол'] == 'Мужской'
data_f = data[mask]
data_m = data[mask2]
data_tmp_f = []
for k,i in enumerate(data_f['Играете ли вы в онлайн-игры?']):      
    if ';' in i:        
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp_f.append(j)                 
    else:
        data_tmp_f.append(i)   
data_tmp_f = pd.Series(data_tmp_f)        
data_tmp_m = []
for k,i in enumerate(data_m['Играете ли вы в онлайн-игры?']):      
    if ';' in i:        
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp_m.append(j)                 
    else:
        data_tmp_m.append(i)   
data_tmp_m = pd.Series(data_tmp_m)
#display(data_tmp_f)
#display(data_tmp_m)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))

count = data_tmp_f.value_counts() 
count2 = data_tmp_m.value_counts()
display(count)
axes[0].pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90,
    explode = [0, 0.01, 0, 0]
)
axes[0].set_title('Играете ли вы в онлайн-игры?\n(Женщины)', fontsize=12);

axes[1].pie(
    count2,
    labels=count2.index,
    autopct='%1.1f%%',
    startangle=90,
    explode = [0.01, 0, 0, 0]
)
axes[1].set_title('Играете ли вы в онлайн-игры?\n(Мужчины)', fontsize=12);
plt.show(pie)
```

### Вывод

Видим, что процент играющих онлайн на компьютере выше среди мужчины (56.2%), чем среди женщины (30.2%). При этом, 45.3% женщин предпочитает играть на телефоне, в то время как среди мужчин этот показатель всего 21.9%

```python
data['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов'].value_counts()
```

```python
# Исключаем респондентов, которые ответили, что не играют в онлайн-игры
mask = data['Играете ли вы в онлайн-игры?'] != 'Нет, не играю'
data_game = data[mask]
#display(data_game['Играете ли вы в онлайн-игры?'].value_counts())
```

```python
data_tmp = []
data_game['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов'] = \
    data_game['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов'].fillna(str(0))
for i in data_game['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов']:    
    i_tmp = i    
    if ';' in i:        
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp.append(j)
    else:
        data_tmp.append(i_tmp)
    
data_game = pd.Series(data_tmp)
display(data_game.value_counts())
```

```python
data_game = data_game.apply(lambda x: 'другое' if (x!='головоломки' and x!='ролевые RPG' and x!='файтинги, стрелялки' and \
                                                        x!='игры с менеджментом фермы, отеля, города' and \
                                                            x!='стратегии' and x!='MMORPG' and x!='гонки' and \
                                                                x!='поиск предметов') else x)
data_game = data_game.apply(lambda x: 'игры с менеджментом\n фермы, отеля, города' if x == 'игры с менеджментом фермы, отеля, города' else x)
data_game = data_game.apply(lambda x: 'файтинги,\n стрелялки' if x == 'файтинги, стрелялки' else x)
data_game = data_game.apply(lambda x: 'поиск\n предметов' if x == 'поиск предметов' else x)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
 
count = data_game.value_counts()

pie = ax.pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90,
    explode = [0, 0, 0, 0.02, 0, 0, 0, 0, 0]
)
title = ax.set_title('В игры каких жанров на ПК вы играете чаще всего?', fontsize=16)
plt.show(pie)
```
