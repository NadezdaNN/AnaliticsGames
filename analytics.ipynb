### Анализ данных, полученных в ходе опроса пользователей интернета, для выявления их отношения к онлайн-играм

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('NewForm.csv')
display(data.head(10))
print('Размерность:', data.shape)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
deaths_count = data['Ваш пол'].value_counts()
pie = ax.pie(
    deaths_count,
    labels=['Женщины', 'Мужчины'],
    autopct='%1.1f%%',
    startangle=90
)
title = ax.set_title('Соотношение респондентов по гендерному признаку', fontsize=16)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
deaths_count = data['Сколько вам лет?'].value_counts() #.sort_values(ascending=False)
pie = ax.pie(
    deaths_count,
    labels=deaths_count.index,
    autopct='%1.1f%%',
    startangle=90
)
title = ax.set_title('Соотношение респондентов по возрасту', fontsize=16)
plt.show(pie)
```

```python
mask = data['Ваш пол'] == 'Женский'
mask2 = data['Ваш пол'] == 'Мужской'
data_tmp = data[mask]
data_tmp2 = data[mask2]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))

count = data_tmp['Сколько вам лет?'].value_counts() 
count2 = data_tmp2['Сколько вам лет?'].value_counts()

axes[0].pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90
)
axes[0].set_title('Соотношение респондентов по возрасту\n среди женщин', fontsize=12);
axes[1].pie(
    count2,
    labels=count2.index,
    autopct='%1.1f%%',
    startangle=90
)

axes[1].set_title('Соотношение респондентов по возрасту\n среди мужчин', fontsize=12);
plt.show(pie)
```

### Вывод.

Видим, что как среди женщин, так и среди мужчин преобладают люди возрастов 25-45 лет (примерно 63% опрошенных), и 18-25 лет (примерно 28% опрошенных). В дальнейшем, нет необходимости отдельно анализировать данные для каждой категории признаков пол и возраст.

```python
data['Играете ли вы в онлайн-игры?'].value_counts()
```

Ответы всего двух респондентов на вопрос: 'Играете ли вы в онлайн-игры?' можно считать неоднозначными: 'Да, на компьютере;Нет, не играю' и 'Да, на телефоне;Нет, не играю'. Будем считать их статистически незначимыми.

```python
data_tmp = []
for i in data['Играете ли вы в онлайн-игры?']:      
    if ';' in i:        
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp.append(j)
    else:
        data_tmp.append(i)    
data_tmp = pd.Series(data_tmp)
display(data_tmp.value_counts())
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
deaths_count = data_tmp.value_counts().sort_values() 
pie = ax.pie(
    deaths_count,
    labels=deaths_count.index,
    autopct='%1.1f%%',
    startangle=90
)
title = ax.set_title('Играете ли вы в онлайн-игры?', fontsize=16)
plt.show(pie)
```

```python
data['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов'].value_counts()
```

```python
# Исключаем респондентов, которые ответили, что не играют в онлайн-игры
mask = data['Играете ли вы в онлайн-игры?'] != 'Нет, не играю'
data_copy = data[mask]
display(data_copy['Играете ли вы в онлайн-игры?'].value_counts())
```

```python
data_tmp = []
data_copy['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов'] = data_copy['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов'].fillna(str(0))
for i in data_copy['В игры каких жанров на ПК вы играете чаще всего, можно выбрать несколько вариантов']:    
    i_tmp = i    
    if ';' in i:        
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp.append(j)
    else:
        data_tmp.append(i_tmp)
    
data_tmp = pd.Series(data_tmp)
display(data_tmp.value_counts())
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes((1, 1, 1, 1))
count = data_tmp.value_counts().head(8)
display(count)
pie = ax.pie(
    count,
    labels=count.index,
    autopct='%1.1f%%',
    startangle=90
)
title = ax.set_title('В игры каких жанров на ПК вы играете чаще всего?', fontsize=16)
plt.show(pie)
```

```python
# Что на ваш взгляд самое важное в игре?
data['Что на ваш взгляд самое важное в игре?'].value_counts()
```

```python
data['Что на ваш взгляд самое важное в игре?'] = data['Что на ваш взгляд самое важное в игре?'].fillna(str('Другое'))
data_tmp = []
for i in data['Что на ваш взгляд самое важное в игре?']:
    #print(i, type(i))
    i_tmp = i
    #print('123=',i)
    if ';' in i:
        #print(i)
        i_tmp = i.split(';')
        for j in i_tmp:
            data_tmp.append(j)
    else:
        data_tmp.append(i_tmp)
    #print(data_tmp)
data_tmp = pd.Series(data_tmp)
display(data_tmp.value_counts())
```
